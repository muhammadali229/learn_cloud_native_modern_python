```python
# create an custom class for exception
class StudentCard():
    def __init__(self, roll_no:int, age:int, name:str)->None:
        if (age < 18) or (age > 68):
            raise StudentAgeError('You are not eligible for this program')
        self.roll_no = roll_no
        self.age = age
        self.name = name
    def get_info(self):
        return {
            'name': self.name,
            'age': self.age,
            'roll_no': self.roll_no
        }
class StudentAgeError(Exception):
        pass

try:
    student_1:StudentCard = StudentCard(1, 19, 'Anus')
    print(student_1.get_info())
except StudentAgeError as e:
    print(e)
try:
    student_1:StudentCard = StudentCard(1, 16, 'Jibran')
    print(student_1.get_info())
except StudentAgeError as e:
    print(e)
```

```python
import PyPDF2

# Open the PDF file
file_path = './PCEP_Muhammad_Ali_Score_Report.pdf'
with open(file_path, 'rb') as file:
    # Create a PDF reader object
    reader = PyPDF2.PdfReader(file)
    
    # Get the number of pages in the PDF
    num_pages = len(reader.pages)
    print(f'Number of pages: {num_pages}')
    
    # Read text from each page
    for page_num in range(num_pages):
        page = reader.pages[page_num]
        text = page.extract_text()
        print(f'Text from page {page_num + 1}:')
        print(text)
        print('-' * 40)
```
