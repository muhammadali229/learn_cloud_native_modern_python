```python
subject_grades['Total'] = 500
subject_grades['obtained'] = subject_grades['s1'] + subject_grades['s2'] + subject_grades['s3'] + subject_grades['s4'] + subject_grades['s5']
```

```python
subject_grades['percentage'] = round((subject_grades['obtained'] / subject_grades['Total']) * 100, 2)
subject_grades.head()
```

```python
# create a custom function grade
def calculate_grade(percentage:float):
    """
    Calculate the grade based on the score according to the Pakistan school grading system.

    Parameters:
    percentage (float): The percentage as a percentage.

    Returns:
    str: The corresponding grade.
    """
    grade = None
    if percentage >= 80:
        grade = 'A+'
    elif percentage >= 70:
        grade = 'A'
    elif percentage >= 60:
        grade = 'B'
    elif percentage >= 50:
        grade = 'C'
    elif percentage >= 40:
        grade = 'D'
    else:
        grade = 'F'
    return grade
subject_grades['grade'] = subject_grades['percentage'].apply(calculate_grade)
```

```python
# apply custom function on multiple cols and return single value means 1 col
def fn_obtained(s1:int, s2:int, s3:int, s4:int, s5:int)->int:
    return s1 + s2 + s3 + s4 + s5
subject_grades['obtained'] = subject_grades[['s1', 's2', 's3', 's4', 's5']].apply(lambda x: fn_obtained(*x), axis=1)
```

```python
def fn_obtained(s1:int, s2:int, s3:int, s4:int, s5:int)->tuple:
    total:int = 500
    obtained:int = s1 + s2 + s3 + s4 + s5
    percentage:float = round((obtained / total) * 100, 2)
    grade:str = calculate_grade(percentage)
    return (
        total,
        obtained,
        percentage,
        grade
    )
subject_grades[['total_m', 'obtained_m', 'percentage_m', 'grade_m']] = subject_grades[['s1', 's2', 's3', 's4', 's5']].apply(lambda x: fn_obtained(*x), axis=1, result_type='expand')
```
